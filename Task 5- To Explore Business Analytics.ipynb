## Creating Subgroups of Profit
**profit** - whose profits are greater than 0  
**neutral** - whose profits are equal to 0  
**loss** - whose profits are less than 0

```python
profit = df[df['Profit'] > 0]
neutral = df[df['Profit'] == 0]
loss = df[df['Profit'] < 0]
```

## Loss with respect to Category

```python
lc_group = loss.groupby('Category')['Profit'].sum().reset_index()
lc_group['Profit'] = lc_group['Profit'].apply(lambda x: x*-1)
print(lc_group)
```

```python
labels = df['Category'].unique()
plt.figure(figsize = (6, 6))
colors = ['gold', 'orange', 'darkorange']

plt.pie(lc_group['Profit'], autopct='%1.1f%%',labels = labels, explode = (0.03,0.03,0.03), colors = colors)

plt.show()
```

**Customer Behaviour:**  
1. Most of the customer choose **Standard Class** over the other three shipping modes. 
2. We got a large amount of customers from the **West** and **East** region.
3. Customers prefer **Office Supplies Products** from us most.
4. **Binders**, **Paper**, **Furnishings**, **Phones** and **Accessories** are the top wanted items from our shop.
<br/>
<br/>

**Profit:**  
1. We got highest sales and profit (**41.6%**) from **Technology**. Interestingly, we got highest profit by selling less number of items from **Technology**.
2. As our customers are mostly from the **West** side, we got most of our profit from there.
3. Our profit got decreased, when we offer any discount on a product. We have got **64.6%** of our profit from the products **without discount** and only **36.4%** of our profit from the products **with discount**
4. Top 10 products which brought most of the profits: **Binders**, **Copiers**, **Phones**, **Accessories**, **Chairs**, **Paper**, **Machines**, **Storage**, **Appliances**, **Furnishings**
5. Top 5 Cities with greater profit: **New York City**, **Los Angeles**, **Seattle**, **San Francisco**, **Detroit**

**Places need to be improved:**  
1. Customers are preferring the **Standard Class** more. We should upgrade the other shipping modes and make them more customer-friendly.

3. We got less customer from the **South**  with a higher loss value. In the **Southern** pert of the country, we will have to be more customer-friendly and come to know there needs.
